```python
import pandas as pd
```

```python
Data=pd.read_csv('blast_dyn_ribo_merged_candidates - blast_dyn_ribo_merged_candidates.csv')
```

```python
from pyensembl import EnsemblRelease
data = EnsemblRelease(release=106)
Data
```

```python
#obj = data.transcript_by_id(transcript_id = "ENST00000369985")
transcriptOf=[]
error_list=[]
for i in Data.index:
    try:
        transcriptOf.append(data.transcript_by_id(transcript_id = str(Data['query'][i])))
    except:
        continue
        
```

```python
data_list=[]
for i in range(len(transcriptOf)):
    data_list.append([transcriptOf[i].transcript_id,transcriptOf[i].transcript_name,transcriptOf[i].gene_id,transcriptOf[i].biotype,
                     transcriptOf[i].contig,transcriptOf[i].start,transcriptOf[i].end,transcriptOf[i].strand,transcriptOf[i].genome,
                     ])
```

```python
data_list_columns=['transcript_id','transcript_name','gene_id','biotype','contig','start','end','strand','genome']
```

```python
df=pd.DataFrame(data_list,columns=data_list_columns)
df
```
