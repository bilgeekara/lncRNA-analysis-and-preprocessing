### Load Data

```python
# load coordinates
with open('transcript_coordinates.json', 'rb') as fp:
    transcript_coordinates = pickle.load(fp)
    
# load disease variance reference dataset
with open('disease_variant_reference.json', 'rb') as fp:
    disease_variant_reference = pickle.load(fp)
```

```python
disease_variant_reference['19']
```

##### Check for each transcript:
* Does it contain disease variants?
* How many disease variants

```python
diseaseVarainceMap = {ID:set() for ID in transcript_coordinates}
```

```python
transcript_coordinates[transcript_id]['contig']
```

```python
for transcript_id in transcript_coordinates.keys():
    chromosome = transcript_coordinates[transcript_id]['contig']
    start = int(transcript_coordinates[transcript_id]['start'])
    end = int(transcript_coordinates[transcript_id]['end'])
    
    # search for disease variants on the given chromosome
    disease_variant_locations = disease_variant_reference[chromosome]
    
    # select varaints that are found within transcript
    for variant_location in disease_variant_locations:
        if variant_location >= start and variant_location <= end:
            diseaseVarainceMap[transcript_id].add(variant_location)
```

```python
diseaseVarainceMap
```

```python
print("Number of transcripts with no disease variants : ", 
      sum([1 for iD in diseaseVarainceMap if diseaseVarainceMap[iD] == set()]))
print("Number of transcripts that contain disease variants : ", 
      sum([1 for iD in diseaseVarainceMap if diseaseVarainceMap[iD] != set()]))
```
